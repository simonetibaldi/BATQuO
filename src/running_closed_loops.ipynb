```python
import time
import random
import datetime
import sys
import numpy 
from utils.parameters import parse_command_line
from utils.bayesian_optimization import *
from utils.default_params import *

np.set_printoptions(precision = 4, suppress = True)
np.set_printoptions(threshold=sys.maxsize)

seed = 12
nwarmup = 10
nbayes = 100
lattice_spacing = 5
verbose_ = True
kernel_choice = 'matern'
discard_percentage = 0.1
load_data = None
```

```python
type_of_graph = "diamond"
depth = 1
shots = 32

####### CREATE BAYES OPT INSTANCE ########
bo_1 = Bayesian_optimization(depth,
                           type_of_graph,
                           lattice_spacing,
                           nwarmup,
                           nbayes,
                           kernel_choice,
                           shots,
                           discard_percentage,
                           seed,
                           verbose_,
                           ) 
bo_1.save_info() 
if load_data==None:    
    bo_1.init_training()
bo_1.run_optimization(load_data)
```
